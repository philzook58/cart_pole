Although, actually as built, the control variable is x of cart. The pendulum is the only dynamical system

```python
m = 1
g = 9.8
l = 0.1
I = m / l * l**3 / 3
L2 = I * thetad ** 2 / 2  + m * (l**2 / 2) * sym.cos(theta) * thetad * xd -  m * g * sym.sin(theta)
eqoM = sym.diff(L2 ,thetad , t) - sym.diff(L2, theta)
```

```python
print(eqoM)
```

```python
thetadd = sym.solve(eqoM, sym.Derivative(thetad, t))[0]
```
